```python
import pandas as pd

#to read csv file
df = pd.read_csv('salaries_by_college_major.csv')
```

```python
#to show first n rows of our dataframe
df.head(6)
```

```python
#to see the number of rows and columns
df.shape
```

```python
#to retrieve the name of columns of the dataframe
df.columns
```

```python
#to retieve last n columns of dataframe
df.tail(5)
```

```python
#returns true if a cell contains anything except a number
df.isna()
```

```python
#dropna removes rows that contains null values
clean_df=df.dropna()
clean_df.tail()
```

```python
#to access data of particular column from dataframe
clean_df['Starting Median Salary']
```

```python
#to find the max value in a particular column
clean_df['Starting Median Salary'].max()
```

```python
#to give us index for the row with the largest value
clean_df['Starting Median Salary'].idxmax()
```

```python
#to see the name of the value of the column in the particular row
clean_df['Undergraduate Major'][43]
```

```python
#to find which major has highest mid career salary
print(clean_df['Mid-Career Median Salary'].max())
print(f"Index for the max mid career salary: {clean_df['Mid-Career Median Salary'].idxmax()}")
clean_df['Undergraduate Major'][8] # we entered row 8 after knowing which row has max mid career salary
```

```python
#to find major with lowest starting and mid career salary
print(clean_df['Starting Median Salary'].min())
clean_df['Undergraduate Major'].loc[clean_df['Starting Median Salary'].idxmin()]
```

```python
#calculating lowest major risk- difference between 10th percentile salary and 90th perecentile salary
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']

#Now to add this column to our existing dataframe use insert() method
clean_df.insert(1, 'Spread', spread_col) # 1 is the column number, Spread is the column name and spead_col is the column to be added
clean_df.head()
```

```python
#to get majors with least difference using sort_values() function
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
#to get majors with highest potential
highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
#to get majors with highest difference in earning
highest_spread = clean_df.sort_values('Spread', ascending=False)
highest_spread[['Undergraduate Major', 'Spread']].head()
```

```python
#to group majors according to groups of business, hass, stem
clean_df.groupby('Group').count()
```

```python
#to display numbers in a particular format
pd.options.display.float_format = '{:,.2f}'.format
```

```python
#to find average salary by group. To skip the non numeric columns use numeric_only=True in mean() function
clean_df.groupby('Group').mean(numeric_only=True)
```
